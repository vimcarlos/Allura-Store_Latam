```python
#Ahora quiero analizar el comportamiento de las ventas en el tiempo, para ello ahora graficaré el numero de articulos vendidos en cada
#categoria por año y vere si es posible por mes, intente por mes pero las gráficas no fueron ilustrativas (demaciada variabilidad), espero que por año sea mejor.



# Notación para el periodo de tiempo 'D' (día), 'S' (semana) o 'M' (mes)
FRECUENCIA = "M"  # intentaré por mes

tiendas = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4,
}


def columna_fecha(df):
    candidatos = [c for c in df.columns if "fecha" in c.lower() or "date" in c.lower()]
    if not candidatos:
        raise ValueError("No se encontró una columna de fecha.")
    return candidatos[0]

for nombre_tienda, df in tiendas.items():
    df = df.copy()
    df.columns = df.columns.str.strip()

    col_fecha = columna_fecha(df)
    col_cat = "Categoría del Producto"

    # Convertir a date-time
    df[col_fecha] = pd.to_datetime(df[col_fecha], errors="coerce", dayfirst=True)
    df = df.dropna(subset=[col_fecha])

    # Extraer el año
    df["Año"] = df[col_fecha].dt.year

    # Conteo de ventas por año y categoría
    ventas_anuales = (
        df.groupby(["Año", col_cat])
          .size()
          .reset_index(name="Ventas")
    )

    # Formato ancho: filas = Año, columnas = Categoría
    piv = (
        ventas_anuales
        .pivot(index="Año", columns=col_cat, values="Ventas")
        .fillna(0)
        .sort_index()
    )
    #Ahora el gráfico
    plt.figure(figsize=(10, 5))
    plt.plot(piv.index, piv.values)
    plt.title(f"{nombre_tienda} – Ventas por categoría por año")
    plt.xlabel("Año")
    plt.ylabel("Número de ventas")
    plt.grid(True, alpha=0.3)

    # Leyendas
    plt.legend(piv.columns, title="Categoría",
               bbox_to_anchor=(1.02, 1), loc="upper left")

    plt.tight_layout()
    plt.show()
```

```python
#Ahora voy a ver el comportameinto de las ventas de cada vendedor por año en cada tienda


DAYFIRST = True
TOP_VENDEDORES = 5  #solo considero los mejores 5 vendedores
def columna_fecha(df):
    candidatos = [c for c in df.columns if "fecha" in c.lower() or "date" in c.lower()]
    if not candidatos:
        raise ValueError("No se encontró columna de fecha (ej: 'Fecha').")
    return candidatos[0]

def columna_vendedor(df):
    candidatos = [c for c in df.columns if "vendedor" in c.lower() or "seller" in c.lower() or "sales" in c.lower()]
    if not candidatos:
        raise ValueError("No se encontró columna de vendedor (ej: 'Vendedor').")
    return candidatos[0]

for nombre_tienda, df in tiendas.items():
    df = df.copy()
    df.columns = df.columns.str.strip()

    col_fecha = columna_fecha(df)
    col_vend = columna_vendedor(df)

    # Convertir a datetime y crear Año
    df[col_fecha] = pd.to_datetime(df[col_fecha], errors="coerce", dayfirst=DAYFIRST)
    df = df.dropna(subset=[col_fecha])
    df["Año"] = df[col_fecha].dt.year

    # Conteo de ventas por año y vendedor
    ventas = (
        df.groupby(["Año", col_vend])
          .size()
          .reset_index(name="Ventas")
    )

    # Solo los mejores vendedores globales (para no saturar la leyenda)
    top_vendedores = (
        ventas.groupby(col_vend)["Ventas"].sum()
              .sort_values(ascending=False)
              .head(TOP_VENDEDORES)
              .index
    )
    ventas = ventas[ventas[col_vend].isin(top_vendedores)]

    # Pivote: filas=Año, columnas=Vendedor
    piv = (ventas.pivot(index="Año", columns=col_vend, values="Ventas")
                 .fillna(0)
                 .sort_index())

 #lineas

    plt.figure(figsize=(10, 5))
    plt.plot(piv.index, piv.values)  # una línea por vendedor
    plt.title(f"{nombre_tienda} – Ventas por vendedor por año (Top {TOP_VENDEDORES})")
    plt.xlabel("Año")
    plt.ylabel("Número de ventas")
    plt.grid(True, alpha=0.3)

    plt.legend(piv.columns, title="Vendedor", bbox_to_anchor=(1.02, 1), loc="upper left")
    plt.tight_layout()
    plt.show()
```

```python
# Me dio curiosidad de ver si con barras se ve mejor que con lineas (ejecute el codigo por tienda)

plt.figure(figsize=(10, 5))
piv.plot(kind="bar", figsize=(10, 5))  # barras agrupadas por año
plt.title(f"{nombre_tienda} – Ventas por vendedor por año (Top {TOP_VENDEDORES})")
plt.xlabel("Año")
plt.ylabel("Número de ventas")
plt.grid(True, axis="y", alpha=0.3)
plt.legend(title="Vendedor", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
# ahora quiero ver como se comporta el numero de cuotas en las que se paga un producto por categorías independientemente de la tienda

def columna_cuotas(df):
    candidatos = [c for c in df.columns if "cantidad" in c.lower() or "cuotas" in c.lower()]
    return candidatos[0]

df = tienda.copy()
df.columns = df.columns.str.strip()

col_cat = "Categoría del Producto"
col_cuotas = columna_cuotas(df)

df[col_cuotas] = pd.to_numeric(df[col_cuotas], errors="coerce")
df = df.dropna(subset=[col_cuotas, col_cat])

# Ordenar categorías por su mediana de cuotas (para que el gráfico sea más legible) porque mean no se ve bien
orden = df.groupby(col_cat)[col_cuotas].median().sort_values().index

data = [df.loc[df[col_cat] == cat, col_cuotas].values for cat in orden]

plt.figure(figsize=(11, 5))
plt.boxplot(data, labels=orden)
plt.xticks(rotation=45, ha="right")
plt.ylabel("Número de cuotas")
plt.xlabel("Categoría del Producto")
plt.title("Distribución del número de cuotas por categoría")
plt.tight_layout()
plt.show()
```

```python
# espero que el gráfico se vea mejor si lo hago por tiendas por separado.

tiendas = {"Tienda 1": tienda, "Tienda 2": tienda2, "Tienda 3": tienda3, "Tienda 4": tienda4}

for nombre, df in tiendas.items():
    df = df.copy()
    df.columns = df.columns.str.strip()

    col_cat = "Categoría del Producto"
    col_cuotas = columna_cuotas(df)

    df[col_cuotas] = pd.to_numeric(df[col_cuotas], errors="coerce")
    df = df.dropna(subset=[col_cuotas, col_cat])

    prom = df.groupby(col_cat)[col_cuotas].median().sort_values(ascending=False)

    plt.figure(figsize=(10, 5))
    plt.bar(prom.index, prom.values)
    plt.xticks(rotation=45, ha="right")
    plt.ylabel("Cuotas promedio")
    plt.xlabel("Categoría del Producto")
    plt.title(f"{nombre} – Cuotas promedio por categoría")
    plt.tight_layout()
    plt.show()
```

```python
# al usar la moda veo que pagar en una sola cuota es lo mas común.

tiendas = {"Tienda 1": tienda, "Tienda 2": tienda2, "Tienda 3": tienda3, "Tienda 4": tienda4}

for nombre, df in tiendas.items():
    df = df.copy()
    df.columns = df.columns.str.strip()

    col_cat = "Categoría del Producto"
    col_cuotas = columna_cuotas(df)

    df[col_cuotas] = pd.to_numeric(df[col_cuotas], errors="coerce")
    df = df.dropna(subset=[col_cuotas, col_cat])

    # Fix: Apply .mode() to each group and select the first mode
    prom = df.groupby(col_cat)[col_cuotas].apply(lambda x: x.mode()[0]).sort_values(ascending=False)

    plt.figure(figsize=(10, 5))
    plt.bar(prom.index, prom.values)
    plt.xticks(rotation=45, ha="right")
    plt.ylabel("Cuotas más frecuente") # Changed label to reflect mode
    plt.xlabel("Categoría del Producto")
    plt.title(f"{nombre} – Cuotas más frecuente por categoría") # Changed title to reflect mode
    plt.tight_layout()
    plt.show()
```

```python
#Para conocer el comprotameinto de las ventas y la localización geográfica de las tiendas
#realziaré graficos de dispercion de las coordenadas de las ventas

import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienas = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4,
}

for nombre, df in tienas.items():
    plt.figure(figsize=(6, 5))

    plt.scatter(
        df["lon"],
        df["lat"],
        alpha=0.5
    )

    plt.xlabel("Longitud")
    plt.ylabel("Latitud")
    plt.title(f"{nombre} – Distribución geográfica de ventas")

    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
```
